```python
import pandas as pd
from sqlalchemy import create_engine

# Define connection details
server = 'AMAL-TAHA'  
database = 'DWH_Ecommerce_F'  
connection_string = (
    f"mssql+pyodbc://@{server}/{database}?"
    "driver=ODBC+Driver+18+for+SQL+Server&"
    "trusted_connection=yes&"
    "TrustServerCertificate=yes"
)

engine = create_engine(connection_string)

try:
    with engine.connect() as conn:
        print("Connection successful!")
except Exception as e:
    print(f"Error: {e}")
```

```python
customers = pd.read_csv('user_dataset.csv')
customers.head()
```

```python
customers = customers.drop(['Unnamed: 4', 'Unnamed: 5'], axis=1)
```

```python
customers = customers.dropna()
```

```python
customers = customers.drop_duplicates(subset='customer_zip_code')
```

```python
customers.shape
```

```python
customers.head()
```

```python
try:
    customers.to_sql('Dim_Customers', engine, if_exists='append', index=False)
    print("customers data loaded successfully!")
except Exception as e:
    print(f"Error: {e}")
```

```python
orders= pd.read_csv('order_dataset.csv')
orders.head()
```

```python
orders['order_approved_date'] = orders['order_approved_date'].fillna(orders['order_approved_date'].mode()[0])
orders['pickup_date'] = orders['pickup_date'].fillna(orders['pickup_date'].mode()[0])
orders['delivered_date'] = orders['delivered_date'].fillna(orders['delivered_date'].mode()[0])
```

```python
orders.isnull().sum()
```

```python
orders = orders.drop(['Unnamed: 8'], axis=1)
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'], format='%m/%d/%Y %H:%M')
orders['order_approved_date'] = pd.to_datetime(orders['order_approved_date'], format='%m/%d/%Y %H:%M')
orders['pickup_date'] = pd.to_datetime(orders['pickup_date'], format='%m/%d/%Y %H:%M')
orders['delivered_date'] = pd.to_datetime(orders['delivered_date'], format='%m/%d/%Y %H:%M')
orders['estimated_time_delivery'] = pd.to_datetime(orders['estimated_time_delivery'], format='%m/%d/%Y %H:%M')
```

```python
order_items= pd.read_csv('order_item_dataset.csv')
order_items.head()
```

```python
order_items.isnull().sum()
```

```python
grouped_order_items = order_items.groupby(['order_id', 'product_id', 'seller_id'], as_index=False).agg({
    'price': 'sum',          
    'shipping_cost': 'sum' 
})
```

```python
grouped_order_items
```

```python
grouped_order_items.isnull().sum()
```

```python
print(customers.columns)
print(sellers.columns)
print(products.columns)
print(payment.columns)
print(feedback.columns)
print(orders.columns)
print(grouped_order_items.columns)
```

```python
# Merging the DataFrames based on 'order_id'
merged_df = products.merge(grouped_order_items, on='product_id', how='left')  # Merge orders with customers
merged_df = merged_df.merge(orders, on='order_id', how='left')  # Merge orders with customers
merged_df = merged_df.merge(payment, on='order_id', how='left')  # Merge orders with customers
merged_df = merged_df.merge(feedback, on='order_id', how='left')  # Merge orders with customers
merged_df = merged_df.merge(customers, on='user_name', how='left')  # Merge orders with customers
merged_df = merged_df.merge(sellers, on='seller_id', how='left')  # Merge orders with customers
print(merged_df.head())
```
